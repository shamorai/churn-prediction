```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
df = pd.read_csv("Customer_Churn.csv")
df.head() #shows top 5 rows and all columns
```

```python
df.info()  #check null values, Data types, Target columns->churn(yes/no)
df.describe()
df.isnull().sum()
```

```python
df['Churn'] = df['Churn'].map({'Yes':1, 'No':0}) # con
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df = df.dropna()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
print(df.columns)
```

```python
X = df[['tenure', 'MonthlyCharges', 'TotalCharges', 'Contract_One year', 'Contract_Two year']]
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python


scaler = StandardScaler()
x_train = scaler.fit_transform(X_train)
x_test = scaler.transform(X_test)

model = RandomForestClassifier(n_estimators=50)
model.fit(X_train, y_train)
```

```python


# Save model
pickle.dump(model, open('model.pkl', 'wb'))

# Save scaler
pickle.dump(scaler, open('scaler.pkl', 'wb'))
```

```python
X_sample = X_test[:100]
```

```python
import shap

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_sample)

shap.summary_plot(shap_values, X_sample)
```
